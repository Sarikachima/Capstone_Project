<H2> Then designed a method to scrap the data from the given website. </H2>

```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup

def scrape_data(url):

    response = requests.get(url, timeout=10)
    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find_all('table')[0]

    rows = table.select('tbody > tr')

    header = [th.text.rstrip() for th in rows[0].find_all('th')]

    with open('output.csv', 'w') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(header)
        for row in rows[1:]:
            data = [th.text.rstrip() for th in row.find_all('td')]
            writer.writerow(data)
```

<H3> Now called the method with the URL as a parameter. Where we can see the data is filled in a dataframe and we can see first five rows with column headers. </H3>

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
scrape_data(url)
df=pd.read_csv("output.csv")
df.head()
```

<H3> Plotting the clusters.</H3>

```python
kclusters=4
map_clusters = folium.Map(location=[toronto_latitude, toronto_longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
i=0
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Explore_df['Latitude'], Explore_df['Longitude'], Explore_df['Neighborhood'], Explore_df['Borough']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    clu=i+1
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[clu],
        fill=True,
        fill_color=rainbow[clu-1],
        fill_opacity=0.7).add_to(map_clusters)
        
map_clusters
```
